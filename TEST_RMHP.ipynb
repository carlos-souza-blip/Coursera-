```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_excel('rmhp_selecao_go.xlsx')
```

```python
dados
```

```python
data = dados[['ano', 'obito', 'icc', 'ieh', 'cid50', 'uso_uti', 'permanencia', 'fxidade']]
data
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_multilabel_classification
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import RidgeClassifierCV
from sklearn.metrics import roc_auc_score, roc_curve, classification_report,\
                            accuracy_score, confusion_matrix, auc
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
modelo = smf.glm(formula= 'obito ~ icc + ieh + uso_uti + permanencia + cid50 + fxidade', data=data,
                 family = sm.families.Binomial()).fit()
print(modelo.summary())
```

```python
model = LogisticRegression(penalty='none', solver='newton-cg')
y = data.obito
x = pd.get_dummies(data[['icc', 'ieh', 'uso_uti', 'permanencia', 'fxidade', 'cid50']])
```

```python
# é necessário adicionar uma constante a matriz X
x = sm.add_constant(x)
# OLS vem de Ordinary Least Squares e o método fit irá treinar o modelo
results = sm.OLS(y, x).fit()
# mostrando as estatísticas do modelo
print(results.summary())
```

```python
import numpy as np
print(results.f_test(np.identity(2)))
```

```python
# mostrando as previsões para o mesmo conjunto passado
print(results.predict(x))
```

```python
# Visualizar os coeficientes encontrados
model.fit(x, y)
print(model.coef_)
```

```python
# Visualizando o R^2
model.score(x, y)
```

```python
# Mostrando as previsões 
model.predict_proba(x)
```
